### Question 1

```python
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
%matplotlib inline
```

```python
np.random.seed(229)
```

```python
mysize = (2000 / (np.array(range(1, 100)))).astype(int)
```

### Chi-Squared Distribution

```python
chisq = [np.random.chisquare(df = 100, size = n) for n in mysize]
chisq_mean = [np.mean(n) for n in chisq]
```

```python
fig = plt.figure(figsize=(10,6)).add_subplot(111)
fig.set_xlabel("Sample Size")
fig.set_ylabel("Sample Mean")
fig.set_title("Chi-Squared Distribution", fontsize = 15)
fig.plot([min(mysize), max(mysize)], [100, 100], 'k')
for n in range(len(chisq_mean)):
    fig.plot(mysize[n], chisq_mean[n], 'o')
```

```python
plt.figure(figsize = (10,6))
plt.hist(chisq_mean)
plt.title("Chi-Squared Distribution", fontsize = 15)
plt.ylabel("Count"); plt.xlabel("Sample Mean")
plt.show()
```

### Poisson Distribution

```python
pois = [np.random.poisson(lam = 100, size = n) for n in mysize]
pois_mean = [np.mean(n) for n in pois]
```

```python
fig = plt.figure(figsize=(10,6)).add_subplot(111)
fig.set_xlabel("Sample Size")
fig.set_ylabel("Sample Mean")
fig.set_title("Poisson Distribution", fontsize = 15)
fig.plot([min(mysize), max(mysize)], [100, 100], 'k')
for n in range(len(pois_mean)):
    fig.plot(mysize[n], pois_mean[n], 'o')
```

```python
plt.figure(figsize = (10,6))
plt.hist(pois_mean)
plt.title("Poisson Distribution", fontsize = 15)
plt.ylabel("Count"); plt.xlabel("Sample Mean")
plt.show()
```

### Binomial Distribution

```python
binom = [np.random.binomial(n = 100, p = 0.3, size = n) for n in mysize]
binom_mean = [np.mean(n) for n in binom]
```

```python
fig = plt.figure(figsize=(10,6)).add_subplot(111)
fig.set_xlabel("Sample Size")
fig.set_ylabel("Sample Mean")
fig.set_title("Binomial Distribution", fontsize = 15)
fig.plot([min(mysize), max(mysize)], [100*0.3, 100*0.3], 'k')
for n in range(len(binom_mean)):
    fig.plot(mysize[n], binom_mean[n], 'o')
```

```python
plt.figure(figsize = (10,6))
plt.hist(binom_mean)
plt.title("Binomial Distribution", fontsize = 15)
plt.ylabel("Count"); plt.xlabel("Sample Mean")
plt.show()
```
