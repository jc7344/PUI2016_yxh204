### Question 1

```python
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
from scipy.stats import norm
%matplotlib inline
```

```python
np.random.seed(229)
```

```python
mysize = (2000 / (np.array(range(1, 100)))).astype(int)
```

### Chi-Squared Distribution

```python
chisq = [np.random.chisquare(df = 100, size = n) for n in mysize]
chisq_mean = [np.mean(n) for n in chisq]
```

```python
fig = plt.figure(figsize=(10,6)).add_subplot(111)
fig.set_xlabel("Sample Size")
fig.set_ylabel("Sample Mean")
fig.set_title("Chi-Squared Distribution", fontsize = 15)
fig.plot([min(mysize), max(mysize)], [100, 100], 'k')
for n in range(len(chisq_mean)):
    fig.plot(mysize[n], chisq_mean[n], 'o')
```

```python
plt.figure(figsize = (10,6))
plt.hist(chisq_mean, normed = True)
plt.title("Chi-Squared Distribution", fontsize = 15)
plt.ylabel("Probability"); plt.xlabel("Sample Mean")

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax)
mu, sigma = norm.fit(chisq_mean)[:2]
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, c = "r")

plt.show()
```

### Explanation:

The law of large numbers states that the sample mean from an experiment will approach the true mean of the population distribution as the number of trials approaches infinity, i.e., $\bar{x} \rightarrow \mu$ as $N \rightarrow \infty$.

In this question, we generated 100 samples of varying sizes $(10<N<2000)$ from each of the 5 different distributions. We then plotted the 100 sample means against the true mean of the population and saw that as the sample size increased, the sample mean tend to vary less and less around the horizontal line representing the true mean of the population.

The histograms that we plotted for the sample means of each of the distributions is a result of the central limit theorem at work. The central limit theorem states that the sample means of sufficiently large number of independent identically distributed random variables will be approximately normally distributed regardless of the underlying distribution that they came from. Here the histograms resemble the normal distribution bell curve.
